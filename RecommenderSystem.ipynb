```python
ratingAvg=pd.DataFrame(ratings.groupby('title') ['rating'].mean())
```

```python
ratingAvg.head()
```

```python
ratingAvg['count']=ratings.groupby('title') ['rating'].count()
ratingAvg.head()
```

```python
mlist=ratings.pivot_table(index='userId',columns='title',values='rating')
```

```python
mlist.head(15)
```

```python
ratings.sort_values('rating',ascending=False)
```

```python
watchedMovie=input('')
```

```python
print(watchedMovie)
```

```python
watchedMovieRating=mlist[watchedMovie]
```

```python
print(watchedMovieRating)
```

```python
similarMovie=mlist.corrwith(watchedMovieRating)
```

```python
similarMovie.head(10)
```

```python
answer=pd.DataFrame(similarMovie,columns=['correlations'])
answer.dropna(inplace=True)
```

```python
answer
```

```python
answer=answer.join(ratingAvg['count'])
```

```python
answer.head()
```

```python
answer[answer['count']>100].sort_values(by='correlations',ascending=False).head()
```
